```python
#################
## LLM Testing ##
#################
from skllm.datasets import get_classification_dataset
from skllm.config import SKLLMConfig
from skllm.models.gpt.classification.zero_shot import ZeroShotGPTClassifier
from skllm.models.gpt.classification.few_shot import FewShotGPTClassifier

from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np

# Configure the credentials
SKLLMConfig.set_openai_key("")
SKLLMConfig.set_openai_org("")
```

```python
## Bring in data
RawData = pd.read_csv("C:/Users/abbal/Documents/Data Science Personal Projects/Py - LLM Classification/ecommerceDataset.csv", header=None).sample(1000)
RawData.columns = ['category', 'text']
```

```python
## Set up data
categories = RawData['category'].unique()
print(categories)

# Splitting the dataset into the Training set and Test set
X = RawData['text']
y = RawData['category']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.95, random_state = 2024)
```

## Zero Shot Classification

```python
# Initialize the model and make the predictions
clf = ZeroShotGPTClassifier(model="gpt-4o-mini")
clf.fit(X_train,y_train)
```

```python
predictions = clf.predict(X_test)
```

```python
pd.crosstab(index = predictions, 
            columns = y_test)
```

## Few Shot Classification

```python
clf_fs = FewShotGPTClassifier(model="gpt-4o-mini")
clf_fs.fit(X_train,y_train)
```

```python
predictions_fs = clf_fs.predict(X_test)
```

```python
pd.crosstab(index = predictions_fs, 
            columns = y_test)
```
